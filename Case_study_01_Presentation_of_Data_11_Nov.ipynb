Do the following operations on ‘cars_data’ dataset:
1. Read the dataset to the python environment.
2. Check for the null values present in the dataset.
3. Plot a bar graph of male vs female buyers participated in the sales.
4. Find the top 5 cars based on their sales price.
5. Find the least 5 cars based on their Resell price.

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the dataset to the python environment.

data = pd.read_csv("cars_data.csv")
```

```python
data.head()
```

```python
#Check for the null values present in the dataset.

data[data.isna().any(axis=1)]
```

```python
#Check for the null values present in the dataset.
data.isnull().sum()
```

```python
#Plot a bar graph of male vs female buyers participated in the sales.
data["Buyer Gender"].value_counts()
```

```python
plt.style.use("classic")
sns.countplot(data["Buyer Gender"])
plt.title("Buyer gender plot")
plt.ylim([4000, 5500])
plt.grid()
```

```python
#Find the top 5 cars based on their sales price.
sale_price_sort= data.sort_values( by =["Sale Price"] , ascending =False)
sale_price_sort.head()
```

```python
#Find the least 5 cars based on their Resell price.

Resale_price_sort= data.sort_values( by =["Resell Price"] , ascending =True)
Resale_price_sort.head()
```
