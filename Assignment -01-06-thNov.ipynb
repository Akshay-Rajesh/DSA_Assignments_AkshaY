# visualizations using the data set

# Bar graph

```python
plt.bar(data["SL"] ,data["Classification"] , color ="r" ,width =0.2)
plt.title("sepal length vs classification")
plt.xlabel("SL")
plt.ylabel("SW")
plt.grid()
plt.plot()
plt.show()
```

# scatter plot

```python
plt.figure(figsize = (6,4) )
plt.scatter(data["PW"] , data["Classification"] , color ="blue" , marker ="*")
plt.xlabel("PW")
plt.ylabel("Classification")
plt.show()
```

# violin plot

```python
sns.violinplot(x =data["SL"] , y = data ["Classification"])
```

# pair plot

```python
sns.pairplot(data)
```

# subplot

```python
plt.subplot(2,1,1)
sns.stripplot(x ="Classification" , y ="PW" , data = data)
plt.subplot(2,1,2)
sns.swarmplot(x ="Classification" , y ="PW" , data = data)
```

# count plot

```python
sns.countplot(data["Classification"])
plt.grid()
```
